# RSNA Bone Age Dataset Audit

Periodic sanity checks:
- CSV schema validation
- Image existence + loadability
- Label distribution checks (train)
- Basic image size stats + sample grid


```python
from pathlib import Path
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

DATA_ROOT = Path(r"C:\Users\hyeon\Documents\miniconda_medimg_env\data\boneage")
TRAIN_CSV = DATA_ROOT / "boneage-training-dataset.csv"
TEST_CSV  = DATA_ROOT / "boneage-test-dataset.csv"
TRAIN_IMG_DIR = DATA_ROOT / "boneage-training-dataset"
TEST_IMG_DIR  = DATA_ROOT / "boneage-test-dataset"

print('DATA_ROOT:', DATA_ROOT)
print('TRAIN_CSV exists:', TRAIN_CSV.exists())
print('TEST_CSV exists :', TEST_CSV.exists())
print('TRAIN_IMG_DIR exists:', TRAIN_IMG_DIR.exists())
print('TEST_IMG_DIR exists :', TEST_IMG_DIR.exists())
```

```python
train_df = pd.read_csv(TRAIN_CSV)
test_df  = pd.read_csv(TEST_CSV)
display(train_df.head())
display(test_df.head())
print('Train shape:', train_df.shape)
print('Test shape :', test_df.shape)
print('Train columns:', list(train_df.columns))
print('Test columns :', list(test_df.columns))
```

```python
# Schema checks
assert set(['id','boneage','male']).issubset(train_df.columns), 'Unexpected train CSV schema'
assert set(['Case ID','Sex']).issubset(test_df.columns), 'Unexpected test CSV schema'

# Label sanity
boneage = train_df['boneage'].astype(float)
print('boneage min/max:', boneage.min(), boneage.max())
print('boneage percentiles:', np.percentile(boneage, [0,1,5,25,50,75,95,99,100]))

plt.figure()
plt.hist(boneage, bins=50)
plt.title('Bone age distribution (months)')
plt.xlabel('months')
plt.ylabel('count')
plt.show()
```

```python
def find_image(img_dir: Path, case_id: int, exts=('png','jpg','jpeg')):
    for ext in exts:
        p = img_dir / f"{int(case_id)}.{ext}"
        if p.exists():
            return p
    hits = list(img_dir.glob(f"{int(case_id)}.*"))
    return hits[0] if hits else None

# Existence audit (sampled train for speed)
train_missing = []
for cid in train_df['id'].astype(int).values[:2000]:
    if find_image(TRAIN_IMG_DIR, cid) is None:
        train_missing.append(cid)

test_missing = []
for cid in test_df['Case ID'].astype(int).values:
    if find_image(TEST_IMG_DIR, cid) is None:
        test_missing.append(cid)

print('Train missing (first 2000):', len(train_missing))
print('Test missing:', len(test_missing))
if train_missing[:10]:
    print('Train missing examples:', train_missing[:10])
if test_missing[:10]:
    print('Test missing examples:', test_missing[:10])
```

```python
# Image loadability + basic size stats + grid (robust)
from pathlib import Path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

from pathlib import Path
from typing import Union

def _normalize_img_dir(img_dir: Union[str, Path]) -> Path:
    img_dir = Path(img_dir)

    # If img_dir contains exactly one child directory with the same name, use the child
    child = img_dir / img_dir.name
    if img_dir.exists() and child.exists() and child.is_dir():
        return child

    return img_dir

def find_image_or_none(img_dir: Path, case_id: int, exts=("png","jpg","jpeg")):
    for ext in exts:
        p = img_dir / f"{int(case_id)}.{ext}"
        if p.exists():
            return p
    hits = list(img_dir.glob(f"{int(case_id)}.*"))
    return hits[0] if hits else None

rng = np.random.default_rng(42)

# Try to sample, but only keep IDs that resolve to an actual file.
candidate_ids = train_df["id"].astype(int).values
rng.shuffle(candidate_ids)

resolved = []
missing = []
for cid in candidate_ids:
    p = find_image_or_none(TRAIN_IMG_DIR, cid)
    if p is None:
        missing.append(cid)
    else:
        resolved.append((cid, p))
    if len(resolved) >= 36:   # target sample size
        break

print(f"Resolved sample images: {len(resolved)}")
print(f"Missing encountered while sampling: {len(missing)}")
if missing:
    print("First 20 missing IDs:", missing[:20])
    print("Expected example filename:", TRAIN_IMG_DIR / f"{missing[0]}.png")

assert len(resolved) > 0, "No images could be resolved. Check TRAIN_IMG_DIR and filename convention."

sizes = []
imgs = []
for cid, p in resolved:
    try:
        im = Image.open(p).convert("L")
    except Exception as e:
        raise RuntimeError(f"Failed to open image for id={cid} at {p}: {e}") from e
    sizes.append(im.size)  # (W,H)
    imgs.append(im)

ws = [s[0] for s in sizes]
hs = [s[1] for s in sizes]
print("Width  min/median/max:", np.min(ws), np.median(ws), np.max(ws))
print("Height min/median/max:", np.min(hs), np.median(hs), np.max(hs))

n = 6
plt.figure(figsize=(12, 12))
for i, im in enumerate(imgs[: n * n]):
    ax = plt.subplot(n, n, i + 1)
    ax.imshow(im, cmap="gray")
    ax.axis("off")
plt.suptitle("Random sample: training images")
plt.show()
```
